```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
flipkart_3 = pd.read_excel('flipkart_3.xlsx')
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="product_category", data=flipkart_3)
```

```python
flipkart_4 = flipkart_3.copy()
```

Merging footwear with beautytoys&more

```python
for i in range(0, 20000):
    if flipkart_4['product_category'][i].find('footwear')!=-1:
        flipkart_4['product_category'][i]='beautytoys&more'
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="product_category", data=flipkart_4)
```

We have alot of class imbalance as expected

First, I will proceed with this imprefection and perform NLP, Visualization and training on this data only
Then I reduce the data such the class distribution is same and train the same model their to see the difference between accuracy

```python
del_col = ['Unnamed: 0']
flipkart_4 = flipkart_4.drop(columns=del_col)
```
